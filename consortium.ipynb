```python
import loica as lc
import matplotlib.pyplot as plt
```

```python
rep_cons = lc.GeneticNetwork()

# Signalling molecules
c4_reg = lc.Regulator(name='C4', degradation_rate=1)
scb2_reg = lc.Regulator(name='SCB2', degradation_rate=1)
c14_reg = lc.Regulator(name='C14', degradation_rate=1)

rep_cons.add_regulator([c4_reg, scb2_reg, c14_reg])

# reporters
mcherry_rep = lc.Reporter(name='mCherry', degradation_rate=1, color='red')
cyan_rep = lc.Reporter(name='CFP', degradation_rate=1, color='blue')
mven_rep = lc.Reporter(name='mVenus', degradation_rate=1, color='yellow')

rep_cons.add_reporter([mcherry_rep, cyan_rep, mven_rep])

# Operators - strains
node_1 = lc.Hill1(name='Strain 1', input=c4_reg, output=[scb2_reg, mcherry_rep], alpha=[10000,0.1], K=10, n=2)
node_2 = lc.Hill1(name='Strain 2', input=scb2_reg, output=[c14_reg, mven_rep], alpha=[10000,0.1], K=10, n=2)
node_3 = lc.Hill1(name='Strain 3', input=c14_reg, output=[c4_reg, cyan_rep], alpha=[10000,0.1], K=10, n=2)


rep_cons.add_operator([node_1, node_2, node_3])
```

```python
plt.figure(figsize=(3.5,3.5), dpi=100)
rep_cons.draw()
```

```python
plt.figure(figsize=(3.3,3.3), dpi=100)
rep_cons.draw(contracted=True)
```
