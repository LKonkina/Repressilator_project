```python
import loica as lc
import matplotlib.pyplot as plt
import networkx as nx
```

```python
cell = lc.GeneticNetwork()

reg1 = lc.Regulator(name='Reg 1', degradation_rate=1)
reg2 = lc.Regulator(name='Reg 2', degradation_rate=1)
reg3 = lc.Regulator(name='Reg 3', degradation_rate=1)

pr1 = lc.Regulator(name="Pr 1")
pr2 = lc.Regulator(name="Pr 2")
pr3 = lc.Regulator(name="Pr 3")
pr4 = lc.Regulator(name="Pr 4")

cell.add_regulator([reg1,reg2, reg3, pr1, pr2, pr3, pr4])

circ1 = lc.Hill1(name='Prom1', input=reg1, output=pr1, alpha=[0.1, 10000], K=10, n=2)
circ2 = lc.Hill1(name='Prom2', input=reg2, output=pr2, alpha=[10000, 0.1], K=10, n=2)
circ3 = lc.Hill1(name='Prom3', input=reg3, output=pr3, alpha=[0.1, 10000], K=10, n=2)
circ4 = lc.Hill1(name='Prom3', input=reg3, output=pr4, alpha=[10000, 0.1], K=10, n=2)

cell.add_operator([circ1, circ2, circ3, circ4])
```

```python
plt.figure(figsize=(3.5,3.5), dpi=200)
cell.draw(pos=nx.planar_layout)
```
